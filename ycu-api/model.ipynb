```python
import pandas as pd
import nltk
import torch
from gramformer import Gramformer

nltk.download('punkt')
```

```python
train_df = pd.read_csv('/workspaces/ycu-gradiation/DataSet/training_set_rel3.tsv',sep='\t', encoding='ISO-8859-1')
train_df.tail()
```

```python
def set_seed(seed):
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(1212)
def gram_former(text):
    gf = Gramformer(models = 1, use_gpu=False) # 1=corrector, 2=detector
    influent_sentences = nltk.sent_tokenize(text)
    gram_score = 0
    missCount = 0
    for influent_sentence in influent_sentences:
        corrected_sentences = gf.correct(influent_sentence, max_candidates=1)
        for corrected_sentence in corrected_sentences:
            missCount = (missCount+gf.highlight(influent_sentence, corrected_sentence[0]).count('</e>'))
    gram_score = missCount/len(text.split())
    return gram_score
```

```python
X = train_df['essay']

Y = []
for essay in X:
    Y.append(gram_former(essay))



```

```python
#線形回帰と非線形回帰
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np

# 最小二乗法　で良い
rgs1 = Lasso() #線形回帰


rgs1.fit(X,Y)
y = rgs1.predict(np.arange(0,10000).reshape(10000,1))

plt.plot(y, label = "First-order equation")
plt.xlabel("domain_Score" ,size = 15) #x軸のタイトル
plt.ylabel("gram_Score", size=15) #y軸のタイトル
plt.scatter(X,  Y,  s=1, c ="r", label = "price_sqftliving")
plt.xlim(0, 10000)
plt.legend()
plt.show()
plt.clf()

```

```python
from sklearn.model_selection import GridSearchCV

rgs = Lasso(normalize=True)

Y_Score = train_df['essay']
X_Score = []
for essay in X_Score:
    X_Score.append(gram_former(essay))


gs_Score = GridSearchCV(rgs, cv = 20 , n_jobs =-1,scoring= "neg_mean_squared_error")
gs_Score.fit(X_Score, Y_Score)
gs_Score.cv_results_
```
